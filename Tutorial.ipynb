### 实际证明product的操作表示张量积！！！

```python
import numpy as np
a = np.random.rand(3)
b = np.random.rand(3)
c = np.random.rand(3)
```

```python
d = np.einsum("i,j,k->ijk",a,b,c)
print("shape of d is; ",d.shape)
print("the output of tensor product is: ",d, end='\n')
```

```python
from itertools import product
res = np.ones((3,3,3))
for i,j,k in product(range(3),range(3),range(3)):
    res[i,j,k] = a[i] * b[j] * c[k]
print('another d caculation result is: ',res)
```

```python
res == d
```

看到一堆True了吧，两重计算方法的结果是完全相同的

### 下面将更清晰的复现代码里的操作

```python
a = np.random.rand(2,3)
b = np.random.rand(2,3)
c = np.random.rand(2,3)
d = np.random.rand(2,3)
res = np.einsum("pi,pj,pk,pm->pijkm",a,b,c,d)
print("result shape is: ",res.shape)
print(res)
```

```python
ano_res = np.zeros((2,3,3,3,3))
bond = 3
for i,j,k,l in product(range(bond),range(bond),range(bond),range(bond)):
    ano_res[:,i,j,k,l] = a[:,i] * b[:,j] * c[:,k] * d[:,l]
print("result shape is: ",ano_res.shape)
print(ano_res)
```

```python
res == ano_res
```

结果一样
